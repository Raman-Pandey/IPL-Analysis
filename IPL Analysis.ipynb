# City

```python
iplcity=ipl['city'].value_counts()
iplcity
```

```python
#graphical representation
plt.figure(figsize=(14,8))
plt.plot(iplcity)
plt.xticks(rotation=90)
plt.title('total number of mathces conducted in cities')
plt.grid(True)
plt.show()
```

```python
venue_ser = ipl['venue'].value_counts()
```

```python
venue_df = pd.DataFrame(columns=['venue', 'matches'])
for items in venue_ser.iteritems():
    temp_df = pd.DataFrame({
        'venue':[items[0]],
        'matches':[items[1]]
    })
    venue_df = venue_df.append(temp_df, ignore_index=True)
```


# Player of the Match

### data of palyer of the match from 2008-2020

```python
ipl['player_of_match'].value_counts() 
```

```python

#ipl['player_of_match'].value_counts().iplot()

df = ipl.copy()
df = df.groupby('player_of_match').filter(lambda x: len(x) > 5)

alt.Chart(df,title="Man Of the Match").mark_bar().encode(alt.X('player_of_match',title = "Player Of the Match"),alt.Y('count()',title = "Total"),color = "player_of_match",tooltip ="player_of_match")
```
